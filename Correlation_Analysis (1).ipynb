### Reading csv & excel files

```python
india = pd.read_csv('Covid19_data_edited.xlsx')
pak = pd.read_excel('Covid19_data_edited.xlsx','Pakistan') 
sl = pd.read_excel('Covid19_data_edited.xlsx','Srilanka') 
ban = pd.read_excel('Covid19_data_edited.xlsx','Bangladesh')
```

# Analysis of India

```python
### Ploting 
sns.heatmap(india.corr())
```

```python
cases = india.dropna()
sns.pairplot(cases, size=1.5)
```

### Adding Correlation Coeffecient to plot

```python
from scipy import stats

def corrfunc(x, y, **kws):
    r, _ = stats.pearsonr(x, y)
    ax = plt.gca()
    ax.annotate('r = {:.2f}'.format(r), xy=(0.1, 1.0), xycoords=ax.transAxes)

pair_plot = sns.pairplot(cases, size=1.5)
pair_plot.map_lower(corrfunc)
pair_plot.map_upper(corrfunc)
```
